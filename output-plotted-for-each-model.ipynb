```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Set Seaborn style
sns.set(style="whitegrid", font_scale=1.2)

# ========================
# 1. BAR GRAPH: Base Val Accuracy Comparison
# ========================
models = ['VGG16', 'Vision Transformer', 'EfficientNet', 'RetNet']
val_accuracies = [0.8533, 0.9133, 0.952, 0.9640]

plt.figure(figsize=(8,6))
sns.barplot(x=models, y=val_accuracies, palette="Blues_d")  # Blues palette for a soft, cool look
plt.title('Validation Accuracy Comparison', fontsize=16, weight='bold')
plt.ylabel('Validation Accuracy', fontsize=14)
plt.ylim(0.7, 1.0)
for index, value in enumerate(val_accuracies):
    plt.text(index, value + 0.01, f"{value:.2f}", ha='center', va='bottom', fontsize=12, weight='bold', color='darkblue')
plt.show()

# ========================
# 2. LINE GRAPH: Loss and Val Accuracy over Epochs
# (Simulated values based on trend, you can replace with real epoch data if available)
# ========================
epochs_vgg16 = np.arange(1, 11)
loss_vgg16 = [1.63, 1.42, 1.29, 1.19, 1.10, 1.01, 1.00, 0.94, 0.86, 0.77]
val_acc_vgg16 = [0.54, 0.56, 0.60, 0.53, 0.65, 0.59, 0.68, 0.77, 0.83, 0.85]

epochs_vit = np.arange(1, 11)
loss_vit = [1.20, 0.71, 0.53, 0.56, 0.42, 0.36, 0.34, 0.32, 0.31, 0.29]
val_acc_vit = [0.77, 0.85, 0.90, 0.89, 0.90, 0.90, 0.91, 0.90, 0.91, 0.91]

epochs_efficientnet = np.arange(1, 11)
loss_eff = [1.61, 1.11, 0.77, 0.61, 0.45, 0.33, 0.30, 0.25, 0.21, 0.17]
val_acc_eff = [0.54, 0.69, 0.49, 0.72, 0.71, 0.69, 0.77, 0.85, 0.96, 0.96]

epochs_retnet = np.arange(1, 11)
loss_retnet = [1.73, 1.56, 1.32, 1.14, 0.89, 0.60, 0.40, 0.33, 0.27, 0.23]
val_acc_retnet = [0.50, 0.76, 0.79, 0.84, 0.84, 0.93, 0.94, 0.95, 0.95, 0.96]

plt.figure(figsize=(14,10))

# VGG16
plt.subplot(2,2,1)
plt.plot(epochs_vgg16, loss_vgg16, label='Loss', marker='o', color='firebrick')
plt.plot(epochs_vgg16, val_acc_vgg16, label='Val Accuracy', marker='o', color='forestgreen')
plt.title('VGG16 Loss and Accuracy', fontsize=14, weight='bold')
plt.xlabel('Epoch', fontsize=12)
plt.ylabel('Value', fontsize=12)
plt.legend(fontsize=12)

# Vision Transformer
plt.subplot(2,2,2)
plt.plot(epochs_vit, loss_vit, label='Loss', marker='o', color='darkorange')
plt.plot(epochs_vit, val_acc_vit, label='Val Accuracy', marker='o', color='mediumblue')
plt.title('Vision Transformer Loss and Accuracy', fontsize=14, weight='bold')
plt.xlabel('Epoch', fontsize=12)
plt.ylabel('Value', fontsize=12)
plt.legend(fontsize=12)

# EfficientNet
plt.subplot(2,2,3)
plt.plot(epochs_efficientnet, loss_eff, label='Loss', marker='o', color='indianred')
plt.plot(epochs_efficientnet, val_acc_eff, label='Val Accuracy', marker='o', color='seagreen')
plt.title('EfficientNet Loss and Accuracy', fontsize=14, weight='bold')
plt.xlabel('Epoch', fontsize=12)
plt.ylabel('Value', fontsize=12)
plt.legend(fontsize=12)

# RetNet
plt.subplot(2,2,4)
plt.plot(epochs_retnet, loss_retnet, label='Loss', marker='o', color='goldenrod')
plt.plot(epochs_retnet, val_acc_retnet, label='Val Accuracy', marker='o', color='mediumslateblue')
plt.title('RetNet Loss and Accuracy', fontsize=14, weight='bold')
plt.xlabel('Epoch', fontsize=12)
plt.ylabel('Value', fontsize=12)
plt.legend(fontsize=12)

plt.tight_layout()
plt.show()

# ========================
# 3. BAR GRAPH: Precision, Recall, F1, Support per model
# ========================
metrics = ['Precision', 'Recall', 'F1-Score', 'Support']

vgg16_scores = [0.90, 1.00, 0.95, 0.83]
vit_scores = [0.94, 0.86, 0.90, 0.85]
efficientnet_scores = [0.91, 0.94, 0.96, 0.94]
retnet_scores = [0.94, 1.00, 0.97, 0.96]

model_scores = [vgg16_scores, vit_scores, efficientnet_scores, retnet_scores]

plt.figure(figsize=(14,10))

for i, scores in enumerate(model_scores):
    plt.subplot(2,2,i+1)
    sns.barplot(x=metrics, y=scores, palette="coolwarm_r")  # Use reversed "coolwarm" for better contrast
    plt.title(f'{models[i]} Metrics', fontsize=14, weight='bold')
    for index, value in enumerate(scores):
        plt.text(index, value + 0.02, f"{value:.2f}", ha='center', va='bottom', fontsize=12, weight='bold', color='darkblue')

plt.tight_layout()
plt.show()
```
